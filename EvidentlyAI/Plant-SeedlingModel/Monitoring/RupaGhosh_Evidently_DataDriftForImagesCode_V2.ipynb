```python
## We will load two sets of images from the two paths below
## data 2 will correspond to the new data and data 1 will refer to original data that the model was trained on
data_2='/content/drive/My Drive/ModelMonitoringBlog/Rupas_Final_Files/Data/Data 2'  ## This contains Shepherds Purse pictures
data_1='/content/drive/My Drive/ModelMonitoringBlog/Rupas_Final_Files/Data/Data1'  ## This contains Sugar Beet
data_1b='/content/drive/My Drive/ModelMonitoringBlog/Rupas_Final_Files/Data/Data1b'  ## This contains Sugar Beet
```

```python
## Initialize the image arrays that we will use subsequently
image_array_a=[]
image_array_b=[]
image_array_a2=[]
```

```python
from PIL import Image, ImageOps
```

```python
#image_array_a=[]

for file in os.listdir(data_1):
    path = data_1+'/' +file
    #print(path)
    temp_image = PIL.Image.open(path)
    temp_image = temp_image.resize((5, 5), Image.ANTIALIAS)
    temp_image=ImageOps.grayscale(temp_image)    # Convert to grayscale
    A = np.array(temp_image)            
    temp_image.show()
    A=A.flatten()
    #print('this is flattenned')
    #print(A)
    image_array_a.append(A)
```

```python
#print (image_array_a)
```

```python
image_array_a2=[]  ## We will read in shepherds Purse

for file in os.listdir(data_1b):
    path = data_1b+'/' +file
    #print(path)
    temp_image = PIL.Image.open(path)
    temp_image = temp_image.resize((5, 5), Image.ANTIALIAS)
    temp_image=ImageOps.grayscale(temp_image)    # Convert to grayscale
    A = np.array(temp_image)            
    #temp_image.show()
    A=A.flatten()
    #print('this is flattenned')
    #print(A)
    image_array_a2.append(A)
```

```python
image_array_b=[]  ## We will read in shepherds Purse

for file in os.listdir(data_2):
    path = data_2+'/' +file
    #print(path)
    temp_image = PIL.Image.open(path)
    temp_image = temp_image.resize((5, 5), Image.ANTIALIAS)
    temp_image=ImageOps.grayscale(temp_image)    # Convert to grayscale
    A = np.array(temp_image)            
    #temp_image.show()
    A=A.flatten()
    #print('this is flattenned')
    #print(A)
    image_array_b.append(A)
```
